## ETL Process

```python
import pandas as pd
```

```python
accident2019_df = pd.read_excel('resources/accident2019.xlsx')
accident2019_df.head()
```

```python
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.LGT_CONDNAME == "Dark - Unknown Lighting"].index)
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.LGT_CONDNAME == "Not Reported"].index)
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.LGT_CONDNAME == "Reported as Unknown"].index)
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.LGT_CONDNAME == "Other"].index)
```

```python
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.WEATHERNAME == "Not Reported"].index)
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.WEATHERNAME == "Reported as Unknown"].index)
accident2019_df = accident2019_df.drop(accident2019_df[accident2019_df.WEATHERNAME == "Other"].index)
```

```python
accident2019_df.to_excel(r'resources/new_accident2019.xlsx', index = False)
```
